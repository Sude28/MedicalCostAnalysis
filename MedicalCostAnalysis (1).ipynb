```python
df_encoded=pd.get_dummies(df, columns=['region'])  #Kategorik sÃ¼tunlarÄ± sayÄ±sal olarak temsil ettik.
df_encoded
```

```python
df_encoded['sex'] = df_encoded['sex'].map({'male': 0, 'female': 1})  # Kategorik sÃ¼tunlarÄ± sayÄ±sal olarak temsil ettik.
df_encoded
```

```python
df_encoded['smoker'] = df_encoded['smoker'].map({'yes': 1, 'no': 0})  # Kategorik sÃ¼tunlarÄ± sayÄ±sal olarak temsil ettik.
df_encoded
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() #Modellerin daha iyi performans gÃ¶stermesi iÃ§in Veri Ã¶lÃ§eklendirme iÅŸlemi yaptÄ±k.
scaled_data = scaler.fit_transform(df_encoded)
scaled_df = pd.DataFrame(scaled_data, columns=df_encoded.columns)
scaled_df
```

```python
y=scaled_df['charges']  #"charges" sÃ¼tununu hedef deÄŸiÅŸken olarak belirliyoruz.
x=scaled_df.drop(['charges'],axis=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split  #Bu fonksiyon, veriyi eÄŸitim ve test kÃ¼melerine bÃ¶lmek iÃ§in kullanÄ±lÄ±r.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) #20'ye 80 olarak bÃ¶leriz
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
models = [
    (DecisionTreeRegressor(), "Decision Tree Regressor"),
    (RandomForestRegressor(), "Random Forest Regressor"),
    (Ridge(alpha=1.0), "Ridge")
]

for model, model_name in models:        # For dÃ¶ngÃ¼sÃ¼ ile bÃ¼tÃ¼n modelleri gezdik
  print(f"Training {model_name}...")

  if isinstance(model,(Ridge)):
    param_grid = {'alpha':[0.01,0.1,1.0,10.0]}
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error') #Grid Search kullanarak modeli optimize ediyoruz
    grid_search.fit(x_train,y_train)
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(x_test)

  elif isinstance(model,DecisionTreeRegressor):
    param_grid = {'max_depth':[None,10,20,30]}
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error') #Grid Search kullanarak modeli optimize ediyoruz
    grid_search.fit(x_train,y_train)
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(x_test)

  elif isinstance(model,RandomForestRegressor):
    param_grid = {'n_estimators': [100, 200, 300], 'max_depth': [None, 10, 20]}
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error') #Grid Search kullanarak modeli optimize ediyoruz
    grid_search.fit(x_train, y_train)
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(x_test)

  mse = mean_squared_error(y_test, y_pred)
  r2 = r2_score(y_test, y_pred)
  print(f"Mean Squared Error for {model_name}: {mse}")
  print(f"R-squared for {model_name}: {r2}")

  print("=" * 50)
  # MSE ne kadar dÃ¼ÅŸÃ¼kse ve R2 ne kadar yÃ¼ksekse, modelin tahminleri o kadar iyi demektir.


```
